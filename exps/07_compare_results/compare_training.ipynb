# Compare training process of CNN, MST+CNN and SMel+CNN (Figure 6)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
import os
font = {'family' : 'sans-serif',
        'size'   : 14}

rc('font', **font)

cnn_folder = "../../sed_endtoend/cnn"
mst_folder = "../05_MST_CNN"
smel_folder = "../06_SMel_CNN"

csvNN1 = os.path.join(cnn_folder,'training.log')
csvNN2 = os.path.join(mst_folder,'training.log')
csvNN3 = os.path.join(smel_folder,'training.log')

log1 = np.genfromtxt (csvNN1, delimiter=",")
log2 = np.genfromtxt (csvNN2, delimiter=",")
log3 = np.genfromtxt (csvNN3, delimiter=",")

log1[np.isnan(log1)] = 0
log2[np.isnan(log2)] = 0
log3[np.isnan(log3)] = 0

fig, axs = plt.subplots( nrows=1, ncols=1 )
#plt.title('F1 variations on the three networks')
plt.plot(log1[:,2][:100],label='CNN',lw=3,color=[0,0,0])
plt.plot(log3[:,2][:100],label='MST+CNN',lw=4,color=[0.5,0.5,0.5])
plt.plot(log2[:,2][:100],label='SMel+CNN',lw=4,color=[0.7,0.7,0.7])
plt.legend(loc=4)
plt.xlabel('epoch number')
plt.ylabel('F1 value')
plt.show()
fig.savefig('F1.eps')   # save the figure to file
plt.close(fig)
```
