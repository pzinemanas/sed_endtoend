# Compare training loss of MST and SMel (Figure 2)

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
font = {'family' : 'sans-serif',
        'size'   : 14}

rc('font', **font)

log = True

csv1 = '../../sed_endtoend/smel/training.log'
csv2 = '../../sed_endtoend/mst/training.log'

log1 = np.genfromtxt (csv1, delimiter=",")
log2 = np.genfromtxt (csv2, delimiter=",")
fig, axs = plt.subplots( nrows=1, ncols=1 )
#plt.title('Loss function values for MST and SMel model')
if log:
    plt.semilogy(log1[:,1][:100],label='SMel model',lw=3,color=[0,0,0])
    plt.semilogy(log2[:,1][:100],label='MST model',lw=3,color=[0.7,0.7,0.7])
    plt.ylim([3*10e-5,0.1])
else:
    plt.plot(log1[:,1][:100],label='SMel model',lw=3,color=[0,0,0])
    plt.plot(log2[:,1][:100],label='MST model',lw=3,color=[0.7,0.7,0.7])
plt.legend()
plt.xlabel('epoch number')
plt.ylabel('loss function value')
plt.show()
fig.savefig('losses.eps')   # save the figure to file
plt.close(fig)
```
